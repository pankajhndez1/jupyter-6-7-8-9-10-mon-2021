# • Exception as an object

# example :

```python
a=10
b=0

try :
    d=a/b         
    print(d)            
    print('inside try block')  
    print()
    
except ZeroDivisionError as obj  :
    print(obj)    # description of the error  will be shown to us i.e  : (  division by zero   )                                  
    print()
    
```

# using multiple exception ,and showing  NameError !!

```python
a=10
b=0

try :
    d=a/g      # deviding 'a' with 'g' which is not defined !!   
    print(d)            
    print('inside try block')  
    print()
    
except ZeroDivisionError as obj  :
    print(obj)    # description of the error  will be shown to us i.e  : (  division by zero   )                                  
    print()
    
except NameError as obj :
    print(obj)            # name error will be raised when the name of variable we are using is not defined !!
    
    
    
```

```python
class MyExceptionError(Exception):
    pass

def myfunc():
    acc_bal=10000
    withdraw_amnt=int(input("enter the amount you want to withdraw !! "))
    balance=acc_bal-withdraw_amnt
    if balance<2000:
        raise MyExceptionError('sorry min 2000 amount should be saved in your account ')
    print(balance)

try:
    myfunc()
    
except MyExceptionError as obj:
    print(obj)
```

# creating file using config the logging System :

```python
from logging import *

basicConfig(filename = 'logfile.log')

debug('this is debug !!')         # this data will not be saved in 'logfile.log'  bcoz level set at WARNING 
info('this is info !!')           # this data will not be saved in 'logfile.log'  bcoz level set at WARNING 
warning(' this is warning !!')    # this data will be saved in 'logfile.log'
error(' this is error !!')        # this data will be saved in 'logfile.log'
critical(" this is critical !!")  # this data will be saved in 'logfile.log'
```

# changing the level to DEBUG :

```python
from logging import *

basicConfig(filename = 'logfile.log' ,level=DEBUG , filemode='w')

debug('this is debug !!')     # this data will be saved in 'logfile.log' # must be shown at o/p but not showing so don't worry
info('this is info !!')       # this data will be saved in 'logfile.log' # must be shown at o/p but not showing so don't worry
warning(' this is warning !!')    # this data will be saved in 'logfile.log'
error(' this is error !!')        # this data will be saved in 'logfile.log'
critical(" this is critical !!")  # this data will be saved in 'logfile.log'
```
