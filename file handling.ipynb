# r+ : Open for reading and then writing

```python
# Read then Write
f = open('pkl.txt', mode='r+')
print(f.tell())      # 0
data = f.read()      # will read upto the cursor at the 19 position !!
print(f.tell())      # 19
f.write('Youtube')   # note : everytime we write the cursor also moves forward !!
print(f.tell())      # after writing in it 26
print(data)          # will read the data from start (print(f.tell()  # 0) data = f.read() to the cursor at 19 !!
print(f.tell())     
```

```python
f=open("pkl.txt",mode="+r")
print(f.tell())
data=f.read()
print(f.tell()) # after reading the whole data here ,the cursor has been set to the last position
data1=f.read()  # now as the cursor has been set to the last position , it will read the blank space 
print(data1)    # and will return the blank space to us 
f.seek(0)       # to set the cursor  to the starting position , now we have to do seek(0)
data2=f.read()
print(data2)
f.close()
```

# w+ : Open for writing and then reading. It will overwrite existing data

```python
# writing and then reading It will overwrite existing data

f = open('pkl.txt', mode='w+')
print(f.tell())     # 0
f.write('Youtube')
print(f.tell())     # 7
data = f.read()    
print(f.tell())    
print(data)         # won't show anything as the cursor is at the end !! 
```

# Pickling

```python
import pickle

class Student(object):
    def __init__(self,name,roll,address):
        self.name=name
        self.roll=roll
        self.address=address
    def disp(self):
        print(f'name of student: {self.name} roll number:{self.roll} the address :{self.address}')
              
stu=Student('pankaj',101,'panchkukla')   # structured data
              
with open("stdnt.bin",mode="wb") as f:
              pickle.dump(stu,f)         # pickcling done here !!
              print("pickling done")
# i.e here we have converted the object of our class into byte stream , and stored it in the binary file !!
```

```python
with open("stdnt.bin",mode="rb") as x:
    obj=pickle.load(x)  # unpickling done here !!
    obj.disp()
```

```python
import pickle

class Student(object):
    def __init__(self,name,roll,address):
        self.name=name
        self.roll=roll
        self.address=address
    def disp(self):
        print(f'name of student: {self.name} roll number:{self.roll} the address :{self.address}')
              
              
with open("stdnt.bin",mode="wb") as f:
              stu=Student('pankaj',101,'panchkukla')   # structured data (object of class)
              stu1=Student("rohit",147,"haldwani")    # structured data (object of class)
              pickle.dump(stu,f)         # (stu) pickcling done here !!  # entering data in form of byte !!
              pickle.dump(stu1,f)        # (stu1) pickcling done here !!
              print("pickling done")
# i.e here we have converted the object of our class into byte stream , and stored it in the binary file !!

with open("stdnt.bin",mode="rb") as x:
    obj=pickle.load(x)  # (stu) unpickling done here !!
    obj1=pickle.load(x) # (stu1) unpickling done here !!
    obj.disp()   # calling (stu) data 
    obj1.disp()  # calling (stu1) data
    print("unpickling done")
```
